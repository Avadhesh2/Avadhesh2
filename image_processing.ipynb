```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/image3.jpg')
plt.figure(figsize=(20, 8))
plt.imshow(img)
```

```python
plt.figure(figsize = (20 ,8))
plt.imshow(grid_RGB)
```
